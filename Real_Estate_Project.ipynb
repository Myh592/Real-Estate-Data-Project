```python
#Dependencies
import pandas as pd
import numpy as np
import requests
import json
```

```python
#Arlington County real estate data
construction_data = "DevelopmentProjectApprovedAndUnderConstruction.csv"
construction_df = pd.read_csv(construction_data)
real_estate_data = "Real_Estate_Sale_History.csv"
realestate_sales_df = pd.read_csv(real_estate_data)
realestate_sales_df.head()
```

```python
#only want dups

realestate_sales_amt = realestate_sales_df[realestate_sales_df.saleAmt > 0]
realestate_sales_dups = realestate_sales_amt[realestate_sales_amt.duplicated(['propertyKey'])]
realestate_sales_dups.sort_values(["propertyKey"], inplace=True)
realestate_sales_dups.reset_index(inplace=True)
realestate_sales_dups.drop(['index'], axis=1, inplace=True)
```

```python
# Make a column for zip codes

realestate_sales_dups["ZipCode"] = ""
realestate_sales_dups.head()
```
